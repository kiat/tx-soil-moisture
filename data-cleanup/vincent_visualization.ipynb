```python
# import libraries
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import ipywidgets as widgets
from ipywidgets import interact
from pathlib import Path
from ipywidgets import VBox, HBox, Button, interactive_output
```

```python
def load_data():
    """
    Load revised final data from CSV.
    
    Args:
        station_id (int): Station number (1-6)
        
    Returns:
        pd.DataFrame
    """
    base_dir = Path("../data-cleanup/raw_merged_data")
    file_path = base_dir / "raw_merged_station_1.csv"
    df = pd.read_csv(file_path, index_col=0, parse_dates=True)
    return df
```

```python
df = load_data()
```

```python
df = df.dropna()
df = df.drop('Flag', axis=1)
```

```python
# Widgets for interface

variable_selector = widgets.Dropdown(
    options=df.columns.tolist()
)

def plot_distribution(variable):
    fig = px.histogram(df, x=variable,
                       title=f"Distributioin of {variable}",
                       nbins=30,
                       marginal="box")

    mean_val = df[variable].mean()
    median_val = df[variable].median()
    
    fig.add_vline(x=mean_val, line_dash="dash", line_color="red",
                  annotation_text="Mean", annotation_position="top right")
    fig.add_vline(x=median_val, line_dash="dash", line_color="green",
                  annotation_text="Median", annotation_position="top right")
    
    fig.update_layout(
        xaxis_title=variable,
        yaxis_title='Count',
        showlegend=True,
        height=600,  
        width=800
    )
    fig.show()
    return fig
    
interactive_plot = interactive_output(plot_distribution, {'variable': variable_selector})
display(VBox([HBox([variable_selector])]), interactive_plot)
    
```
