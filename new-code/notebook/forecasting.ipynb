```python
# %pip install -r ../requirements.txt --quiet
# IMPORT LIBRARIES (MUST RUN FIRST)
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# GLOBAL VARIABLES (PERSIST ACROSS CELLS)
DATA_FOLDER = "../processed-data/Revised_Final_Data"
STATION_DATA = {}  # Stores all stations' data
SELECTED_STATION = None
DF = None  # Active station DataFrame

# FUNCTION TO LOAD DATA
def load_station_data(folder_path):
    global STATION_DATA
    station_files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]
    
    for file in station_files:
        station_name = file.replace("_Revised_Final_Data.csv", "")
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path, parse_dates=["Unnamed: 0"], index_col="Unnamed: 0")
        STATION_DATA[station_name] = df
    
    print(f"✅ Loaded data from {len(STATION_DATA)} stations.")

# LOAD DATA ONCE
load_station_data(DATA_FOLDER)
```
