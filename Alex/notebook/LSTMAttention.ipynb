```python
import numpy as np
import pandas as pd

def engineer_data(dfs, boolean):
    # Constants for day and year calculations
    day = 24 * 60 * 60
    year = 365.2425 * day

    for station, df in dfs.items():
        # Drop rows with NaN values to ensure clean data
        df = df.dropna()

        # Extract windspeed and wind direction
        wv = df['Windspeed']
        wd_rad = np.deg2rad(df['Winddirection'])  # Convert degrees to radians

        # Ensure the index is in datetime format
        df.index = pd.to_datetime(df.index)

        # Convert timestamp index to seconds
        timestamp_s = df.index.map(pd.Timestamp.timestamp)

        # Extract latitude and longitude
        lat = np.deg2rad(df['Latitude'])  # Convert latitude to radians
        lon = np.deg2rad(df['Longitude'])  # Convert longitude to radians

        # Engineer wind components

        # print("HELLO", wv * np.cos(wd_rad))
        # print("AHHHHH",wd_rad)

        if boolean:
          df['Wx'] = wv * np.cos(wd_rad)
          df['Wy'] = wv * np.sin(wd_rad)

        # Engineer time-based cyclic features
        df['Day sin'] = np.sin(timestamp_s * (2 * np.pi / day))
        df['Day cos'] = np.cos(timestamp_s * (2 * np.pi / day))
        df['Year sin'] = np.sin(timestamp_s * (2 * np.pi / year))
        df['Year cos'] = np.cos(timestamp_s * (2 * np.pi / year))

        # Engineer Cartesian coordinates for latitude and longitude
        df['x_cord'] = np.cos(lat) * np.cos(lon)
        df['y_cord'] = np.cos(lat) * np.sin(lon)
        df['z_cord'] = np.sin(lat)

        # Update the DataFrame in the dictionary
        dfs[station] = df

    return dfs



```

```python
def scale_data(dfs):
    for station, df in dfs.items():
        cur_df = df.copy()
        d_sin = cur_df.pop("Day sin")
        d_cos = cur_df.pop("Day cos")
        y_sin = cur_df.pop("Year sin")
        y_cos = cur_df.pop("Year cos")
        x = cur_df.pop("x_cord")
        y = cur_df.pop("y_cord")
        z = cur_df.pop("z_cord")
        scaler = MinMaxScaler()
        scaled_df = pd.DataFrame(data=scaler.fit_transform(cur_df), columns=cur_df.columns, index=cur_df.index)
        scaled_df["Day sin"] = d_sin.values
        scaled_df["Day cos"] = d_cos.values
        scaled_df["Year sin"] = y_sin.values
        scaled_df["Year cos"] = y_cos.values
        scaled_df["x_cord"] = x.values
        scaled_df["y_cord"] = y.values
        scaled_df["z_cord"] = z.values
        dfs[station] = scaled_df
```

```python
#Definitions
TARGET_COL = "SWC_5"
TRAIN_SPLIT = 0.7
VAL_SPLIT = 0.2
WINDOW_SIZE = 24 * 7
SHIFT_AMT = 10
PAT = 3
MAX_EPOCHS = 25
```

```python
def split(df, target_col=TARGET_COL, train_split = TRAIN_SPLIT, val_split = VAL_SPLIT):
  print(df.columns)
  target_idx = df.columns.get_loc(target_col)
  train_set = df[ : int(len(df) * train_split)].values
  val_set = df[int(len(df) * train_split) : int(len(df) * (train_split + val_split))].values
  test_set = df[int(len(df) * (train_split + val_split)) :].values
  return (train_set, val_set, test_set, target_idx)
```

```python
def generate_windows(data, window_size=24, shift=24, target_idx=0):
    labels = data[:, target_idx]

    X = []
    y = []
    for i in range(len(data) - window_size - shift):
        # get window based on input width
        window = data[i : i + window_size]
        window_label = labels[i + window_size + shift]

        X.append(window)
        y.append(window_label)

    # in new dataset, each element is a data window, and window label is single value
    return np.array(X), np.array(y)


# given data and its labels, divide the data into batches of size batch_size
def generate_batches(X, y, batch_size=32):
    # divides data into batches, drops any remainder batches smaller than specified batch size.
    # allows models to run with any batch size
    tf_dataset = tf.data.Dataset.from_tensor_slices((X, y))
    tf_dataset = tf_dataset.repeat().batch(batch_size=batch_size, drop_remainder=True)

    # tf_dataset repeats indefinitely, need to compute number of step updates to complete 1 epoch
    steps_per_epoch = len(X) // batch_size

    return (tf_dataset, steps_per_epoch)
```

```python
def preprocess_data(df):
    # data cleaning and feature engineering
    train_set, val_set, test_set, target_idx = split(df)


    # create window data for each dataset
    X_train, y_train = generate_windows(train_set, window_size=WINDOW_SIZE, shift=SHIFT_AMT, target_idx=target_idx)
    X_val, y_val = generate_windows(val_set, window_size=WINDOW_SIZE, shift=SHIFT_AMT, target_idx=target_idx)
    X_test, y_test = generate_windows(test_set, window_size=WINDOW_SIZE, shift=SHIFT_AMT, target_idx=target_idx)

    return (X_train, y_train, X_val, y_val, X_test, y_test)
```

```python
cur_df = dfs["Station1"]

BATCH_SIZE = 128
X_train, y_train, X_val, y_val, X_test, y_test = preprocess_data(cur_df)

# divide each dataset into batched version to feed to models
train_dataset, train_steps = generate_batches(X_train, y_train, batch_size=BATCH_SIZE)
val_dataset, val_steps = generate_batches(X_val, y_val, batch_size=BATCH_SIZE)
test_dataset, test_steps = generate_batches(X_test, y_test, batch_size=BATCH_SIZE)
```

```python
def compile_and_fit(model, data, steps_per_epoch, val_data, val_steps, model_name='model/', patience=3, max_epochs=MAX_EPOCHS, batch_size=32):
    # stop running epochs if the loss stops improving for patience number of epochs
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=patience, mode='min')

    # store the best model on disk to be loaded later without having to re-fit
    # allows you to load models from disc
    ckpt = tf.keras.callbacks.ModelCheckpoint(model_name + ".keras", save_best_only=True)

    model.compile(loss=tf.keras.losses.MeanSquaredError(),
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  metrics=[tf.keras.metrics.MeanAbsoluteError(), tf.keras.metrics.MeanSquaredError(), tf.keras.metrics.MeanAbsolutePercentageError()])

    history = model.fit(data,
                        epochs=max_epochs,
                        callbacks=[ckpt, early_stopping],
                        validation_data=val_data,
                        validation_steps=val_steps,
                        shuffle=False,
                        batch_size=batch_size,
                        steps_per_epoch = steps_per_epoch)

    return history
```

```python
preds = {}

def plot_single_pred(model, name, dataset, data_steps, y, batch_size=32):
    forecast = model.predict(dataset, batch_size=batch_size, steps=data_steps)
    preds[name] = forecast
    if len(forecast.shape) == 3:
        print("asd")
        forecast = forecast[:, 0, 0]
    elif len(forecast.shape) == 2:
        forecast = forecast[:, 0]

    plt.figure(figsize=(10, 6))
    plot_data = {"Predictions": forecast, "Actual": y}

    plt.plot(plot_data["Actual"])
    plt.plot(plot_data["Predictions"])

    plt.legend(("Actual", "Predictions"))

    return plot_data
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, Input, LSTM, Dense
from tensorflow.keras.models import Model

# Scaled Dot-Product Attention Layer
class ScaledDotProductAttention(Layer):
    def __init__(self, **kwargs):
        super(ScaledDotProductAttention, self).__init__(**kwargs)

    def call(self, Q, K, V):
        # Compute the dot products between Q and K^T
        d_k = tf.cast(tf.shape(K)[-1], tf.float32)  # dimension of the key
        scores = tf.matmul(Q, K, transpose_b=True) / tf.sqrt(d_k)  # scale by sqrt(d_k)

        # Apply the softmax function to get the attention weights
        weights = tf.nn.softmax(scores, axis=-1)

        # Multiply the weights by the values (V) to get the attention output
        attention_output = tf.matmul(weights, V)
        return attention_output

# LSTM model with Scaled Dot-Product Attention mechanism
def build_lstm_with_scaled_dot_product_attention(input_shape, d_k, d_v):
    inputs = Input(shape=input_shape)
    lstm_out = LSTM(128, return_sequences=True)(inputs)
    lstm_out = LSTM(64, return_sequences=True)(lstm_out)
    lstm_out = LSTM(32, return_sequences=True)(lstm_out)

    # Compute Q, K, and V
    Q = Dense(d_k)(lstm_out)  # Query
    K = Dense(d_k)(lstm_out)  # Key
    V = Dense(d_v)(lstm_out)  # Value

    # Apply scaled dot-product attention
    attention_output = ScaledDotProductAttention()(Q, K, V)

    # Flatten the output and add Dense layers for prediction
    dense = Dense(32, activation='relu')(attention_output)
    output = Dense(1, activation='tanh')(dense)

    # Define and compile the model
    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss='mse',
                  metrics=['mae', 'mse'])

    return model

# Set dimensions for Q, K, V
d_k = 64  # Dimension of Q and K
d_v = 64  # Dimension of V

# Build and compile the model with scaled dot-product attention
attention_lstm_model = build_lstm_with_scaled_dot_product_attention(X_train.shape[1:], d_k, d_v)

# Train the model
attention_lstm_history = compile_and_fit(attention_lstm_model, train_dataset, train_steps, val_dataset, val_steps,
                                         batch_size=BATCH_SIZE, model_name="scaledDotProductAttentionLSTM", patience=PAT)

# Evaluate the model performance
val_performance["scaledDotProductAttentionLSTM"] = attention_lstm_model.evaluate(val_dataset, steps=val_steps, batch_size=BATCH_SIZE, verbose=1)
performance["scaledDotProductAttentionLSTM"] = attention_lstm_model.evaluate(test_dataset, steps=test_steps, batch_size=BATCH_SIZE, verbose=0)

# Plotting the predictions
plot_data = plot_single_pred(attention_lstm_model, "Scaled Dot-Product Attention LSTM", test_dataset, test_steps, y_test, batch_size=BATCH_SIZE)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, Input, LSTM, Dense, Concatenate
from tensorflow.keras.models import Model

# Multi-Head Attention Layer
class MultiHeadAttention(Layer):
    def __init__(self, num_heads, d_k, d_v, **kwargs):
        super(MultiHeadAttention, self).__init__(**kwargs)
        self.num_heads = num_heads
        self.d_k = d_k
        self.d_v = d_v

    def build(self, input_shape):
        # Define weight matrices for Q, K, V for each head
        self.W_Q = [self.add_weight(name=f"W_Q_{i}", shape=(input_shape[-1], self.d_k),
                                    initializer="random_normal", trainable=True) for i in range(self.num_heads)]
        self.W_K = [self.add_weight(name=f"W_K_{i}", shape=(input_shape[-1], self.d_k),
                                    initializer="random_normal", trainable=True) for i in range(self.num_heads)]
        self.W_V = [self.add_weight(name=f"W_V_{i}", shape=(input_shape[-1], self.d_v),
                                    initializer="random_normal", trainable=True) for i in range(self.num_heads)]
        # Output weight matrix
        self.W_O = self.add_weight(name="W_O", shape=(self.num_heads * self.d_v, input_shape[-1]),
                                   initializer="random_normal", trainable=True)

    def call(self, inputs):
        # Compute Q, K, V for each head
        multi_head_outputs = []
        for i in range(self.num_heads):
            Q = tf.matmul(inputs, self.W_Q[i])
            K = tf.matmul(inputs, self.W_K[i])
            V = tf.matmul(inputs, self.W_V[i])

            # Scaled dot-product attention
            d_k = tf.cast(tf.shape(K)[-1], tf.float32)
            scores = tf.matmul(Q, K, transpose_b=True) / tf.sqrt(d_k)
            weights = tf.nn.softmax(scores, axis=-1)
            attention_output = tf.matmul(weights, V)

            multi_head_outputs.append(attention_output)

        # Concatenate the outputs from all heads
        concatenated_outputs = Concatenate()(multi_head_outputs)

        # Apply the final linear transformation
        output = tf.matmul(concatenated_outputs, self.W_O)
        return output

# LSTM model with Multi-Head Attention mechanism
def build_lstm_with_multi_head_attention(input_shape, num_heads, d_k, d_v):
    inputs = Input(shape=input_shape)
    lstm_out = LSTM(128, return_sequences=True)(inputs)
    lstm_out = LSTM(64, return_sequences=True)(lstm_out)
    lstm_out = LSTM(32, return_sequences=True)(lstm_out)

    # Apply Multi-Head Attention
    attention_output = MultiHeadAttention(num_heads=num_heads, d_k=d_k, d_v=d_v)(lstm_out)

    # Flatten the output and add Dense layers for prediction
    dense = Dense(32, activation='relu')(attention_output)
    output = Dense(1, activation='tanh')(dense)

    # Define and compile the model
    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss='mse',
                  metrics=['mae', 'mse'])

    return model

# Set parameters for multi-head attention
num_heads = 4  # Number of attention heads
d_k = 32       # Dimension of each head for Q and K
d_v = 32       # Dimension of each head for V

# Build and compile the model with multi-head attention
multi_head_attention_lstm_model = build_lstm_with_multi_head_attention(X_train.shape[1:], num_heads, d_k, d_v)

# Train the model
multi_head_attention_history = compile_and_fit(multi_head_attention_lstm_model, train_dataset, train_steps, val_dataset, val_steps,
                                               batch_size=BATCH_SIZE, model_name="multiHeadAttentionLSTM", patience=PAT)

# Evaluate the model performance
val_performance["multiHeadAttentionLSTM"] = multi_head_attention_lstm_model.evaluate(val_dataset, steps=val_steps, batch_size=BATCH_SIZE, verbose=1)
performance["multiHeadAttentionLSTM"] = multi_head_attention_lstm_model.evaluate(test_dataset, steps=test_steps, batch_size=BATCH_SIZE, verbose=0)

# Plotting the predictions
plot_data = plot_single_pred(multi_head_attention_lstm_model, "Multi-Head Attention LSTM", test_dataset, test_steps, y_test, batch_size=BATCH_SIZE)
```

```python
from tensorflow.keras.layers import Dropout, LayerNormalization, Add

# Updated LSTM model with Multi-Head Attention, Dropout, and Residual Connections
def build_improved_lstm_with_multi_head_attention(input_shape, num_heads, d_k, d_v):
    inputs = Input(shape=input_shape)
    lstm_out = LSTM(128, return_sequences=True)(inputs)
    lstm_out = LSTM(64, return_sequences=True)(lstm_out)
    lstm_out = LSTM(32, return_sequences=True)(lstm_out)

    # Apply Multi-Head Attention
    attention_output = MultiHeadAttention(num_heads=num_heads, d_k=d_k, d_v=d_v)(lstm_out)

    # Add residual connection and normalization
    attention_output = Add()([attention_output, lstm_out])  # Residual connection
    attention_output = LayerNormalization()(attention_output)

    # Apply dropout to avoid overfitting

    # Flatten the output and add Dense layers for prediction
    dense = Dense(32, activation='relu')(attention_output)
    output = Dense(1, activation='tanh')(dense)

    # Define and compile the model
    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss='mse',
                  metrics=['mae', 'mse'])

    return model

# Build the improved model
improved_attention_lstm_model = build_improved_lstm_with_multi_head_attention(X_train.shape[1:], num_heads, d_k, d_v)

# Train the improved model
improved_attention_history = compile_and_fit(improved_attention_lstm_model, train_dataset, train_steps, val_dataset, val_steps,
                                             batch_size=BATCH_SIZE, model_name="improvedMultiHeadAttentionLSTM", patience=PAT)

# Evaluate the model performance
val_performance["improvedMultiHeadAttentionLSTM"] = improved_attention_lstm_model.evaluate(val_dataset, steps=val_steps, batch_size=BATCH_SIZE, verbose=1)
performance["improvedMultiHeadAttentionLSTM"] = improved_attention_lstm_model.evaluate(test_dataset, steps=test_steps, batch_size=BATCH_SIZE, verbose=0)

# Plotting the predictions
plot_data = plot_single_pred(improved_attention_lstm_model, "Improved Multi-Head Attention LSTM", test_dataset, test_steps, y_test, batch_size=BATCH_SIZE)
```
