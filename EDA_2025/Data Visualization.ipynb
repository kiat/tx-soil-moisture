## 1: Import Libraries and Define Data Loading Functions

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import ipywidgets as widgets
from ipywidgets import interact
from pathlib import Path
from ipywidgets import VBox, HBox, Button, interactive_output
import webbrowser


def load_cleaned_data(data_type, station_id, prefer_filled_soil=True):
    """
    Load cleaned soil or meteorological data.

    When ``data_type`` is "soil" and ``prefer_filled_soil`` is True, attempt to
    load the gap-filled series produced by VeryLongGaps.py
    (``Model_2025/output/StationX_filled_verylonggaps.csv``). Fall back to the
    original cleaned files if the filled CSV is missing so earlier workflows
    keep working.
    """
    if data_type == "soil" and prefer_filled_soil:
        filled_path = Path("../Model_2025/output") / f"Station{station_id}_filled_verylonggaps.csv"
        if filled_path.exists():
            df = pd.read_csv(filled_path, index_col=0, parse_dates=True)
            return df

    base_dir = Path(f"./cleaned_{data_type}_data")
    file_prefix = "SM" if data_type == "soil" else "MET"
    file_path = base_dir / f"{file_prefix}_{station_id}_cleaned.csv"
    df = pd.read_csv(file_path, index_col=0, parse_dates=True)
    return df
```

## 2. Dynamic Visualization for Soil Data

### 2.1 Soil Moisture Interactive Plot

```python
def plot_soil_moisture_interactive(station_id=1, year=2015, month=1):
    # Load soil data for the selected station
    df = load_cleaned_data('soil', station_id)
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year and month
    df_filtered = df[(df.index.year == year) & (df.index.month == month)]
    
    if df_filtered.empty:
        print(f"No soil moisture data available for Station {station_id} in {year}-{month:02d}")
        return
    
    # Define the soil moisture columns and colors
    swc_cols = ['SWC_5', 'SWC_10', 'SWC_20', 'SWC_50']
    colors = {
        'SWC_5': '#4682B4',  # blue
        'SWC_10': '#FFA500',  # orange
        'SWC_20': '#3CB371',  # green
        'SWC_50': '#DC143C'   # red
    }
    
    # Create the Plotly figure
    fig = go.Figure()
    for col in swc_cols:
        fig.add_trace(go.Scatter(
            x=df_filtered.index,
            y=df_filtered[col],
            mode='lines',
            name=col,
            line=dict(color=colors[col])
        ))
    
    fig.update_layout(
        title=f"Soil Moisture for Station {station_id} in {year}-{month:02d}",
        xaxis_title="Date",
        yaxis_title="Soil Moisture (m³/m³)",
        width=1000,
        height=500
    )
    fig.show()
```

```python
# Interactive control for soil moisture
interact(plot_soil_moisture_interactive,
         station_id=widgets.IntSlider(min=1, max=6, step=1, value=1),
         year=widgets.Dropdown(options=list(range(2015, 2022)), value=2015),
         month=widgets.Dropdown(options=list(range(1, 13)), value=1));
```

## 3. Dynamic Visualization for Meteorological Data

### 3.1 Meteorological Variable Interactive Plot

```python
def plot_met_variable_interactive(station_id, variable='Tair', year=2015, month=1):
    # Load meteorological data for the selected station
    df = load_cleaned_data('met', station_id)
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year and month
    df_filtered = df[(df.index.year == year) & (df.index.month == month)]
    
    if df_filtered.empty:
        print(f"No meteorological data available for Station {station_id} in {year}-{month:02d}")
        return
    
    # Create the Plotly figure
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df_filtered.index,
        y=df_filtered[variable],
        mode='lines',
        name=variable
    ))
    
    fig.update_layout(
        title=f"{variable} for Station {station_id} in {year}-{month:02d}",
        xaxis_title="Date",
        yaxis_title=variable,
        width=1000,
        height=500
    )
    fig.show()
```

```python
# Interactive control for meteorological data
interact(plot_met_variable_interactive,
         station_id=widgets.IntSlider(min=1, max=6, step=1, value=1),
         variable=widgets.Dropdown(options=['Tair', 'RH', 'Wind speed', 'Srad'], value='Tair'),
         year=widgets.Dropdown(options=list(range(2015, 2022)), value=2015),
         month=widgets.Dropdown(options=list(range(1, 13)), value=1));
```

```python
def plot_met_variable_yearly(station_id=1, variable='Tair', year=2015):
    """
    Plot the full-year time series for a specified MET variable.

    Args:
        station_id (int): The MET station number.
        variable (str): The variable to plot (e.g., 'Tair', 'RH', 'Wind speed', 'Srad').
        year (int): The selected year.

    Returns:
        fig (plotly.graph_objects.Figure): The generated Plotly figure.
    """
    # Load the MET data for the selected station
    df = load_cleaned_data('met', station_id)
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year
    df_filtered = df[df.index.year == year]
    
    if df_filtered.empty:
        print(f"No {variable} data available for MET Station {station_id} in {year}")
        return go.Figure()  # Return an empty figure if there's no data

    # Create the Plotly figure for the selected MET variable
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df_filtered.index,
        y=df_filtered[variable],
        mode='lines',
        name=variable,
        line=dict(width=2)
    ))
    
    fig.update_layout(
        title=f"{variable} for MET Station {station_id} in {year}",
        xaxis_title="Date",
        yaxis_title=variable,
        width=1000,
        height=500
    )
    fig.show()
```

### 4. Missing value visualization

```python
# load merged data
def load_merged_data(station_id):
    """
    Load merged soil and meteorological data for a given station.
    
    Args:
        station_id (int): Station number (1-6)
        
    Returns:
        pd.DataFrame: Merged data with a datetime index.
    """
    file_path = Path(f"./merged_data/merged_station_{station_id}.csv")
    df = pd.read_csv(file_path, index_col=0, parse_dates=True)
    return df
```

```python
# Bar chart visualization of missing values
def plot_missing_values_bar(station_id, start_date="2015-01-01", end_date="2021-12-31"):
    df = load_merged_data(station_id)
    df = df.loc[start_date:end_date]
    
    missing_percent = df.isnull().mean() * 100
    
    fig = px.bar(
        x=missing_percent.index,
        y=missing_percent.values,
        labels={"x": "Feature", "y": "Missing Percentage (%)"},
        title=f"Missing Value Percentage per Feature for Station {station_id}"
    )
    fig.show()
```

```python
# Heatmap visualization of missing values over time
def plot_missing_values_heatmap(station_id, start_date="2015-01-01", end_date="2021-12-31"):
    df = load_merged_data(station_id)
    df = df.loc[start_date:end_date]
    
    # Create a binary mask for missing values (1 for missing, 0 for present)
    mask = df.isnull().astype(int)
    
    # Resample daily and compute the average (percentage of missing values per day)
    mask_daily = mask.resample('D').mean() * 100
    
    fig = px.imshow(
        mask_daily.T,
        labels={"x": "Date", "y": "Feature", "color": "Missing (%)"},
        aspect="auto",
        title=f"Daily Missing Value Percentage Heatmap for Station {station_id}"
    )
    fig.update_xaxes(side="top")
    fig.show()
```

```python
# Choose between bar chart and heatmap
def visualize_missing_data_interactive(station_id, start_date, end_date, plot_type):
    if plot_type == "Bar":
        plot_missing_values_bar(station_id, start_date, end_date)
    elif plot_type == "Heatmap":
        plot_missing_values_heatmap(station_id, start_date, end_date)
    else:
        print("Please select a valid plot type.")
```

```python
# Station selection
station_dropdown = widgets.Dropdown(
    options=[1, 2, 3, 4, 5, 6],
    value=1,
    description='Station:',
    disabled=False,
)

# Create interactive controls
interact(
    visualize_missing_data_interactive,
    station_id=station_dropdown,
    start_date=widgets.Text(value="2015-01-01", description="Start Date"),
    end_date=widgets.Text(value="2021-12-31", description="End Date"),
    plot_type=widgets.Dropdown(options=["Bar", "Heatmap"], value="Bar", description="Plot Type")
);
```

## Integrated interactive interface

```python
def update_plot(plot_type, station_id, year, month, met_variable, time_resolution):
    if plot_type == "Soil Moisture":
        if time_resolution == "Monthly":
            fig = plot_soil_moisture_interactive(station_id, year, month)
        else:
            fig = plot_soil_moisture_yearly(station_id, year)
    elif plot_type == "Soil Temperature":
        if time_resolution == "Monthly":
            fig = plot_soil_temperature_interactive(station_id, year, month)
        else:
            fig = plot_soil_temperature_yearly(station_id, year)
    elif plot_type == "MET Variable":
        if time_resolution == "Monthly":
            fig = plot_met_variable_interactive(station_id, met_variable, year, month)
        else:
            fig = plot_met_variable_yearly(station_id, met_variable, year)
    else:
        fig = go.Figure()
    return fig

# Create widgets for the interface

# Plot Type selector
plot_type_selector = widgets.Dropdown(
    options=["Soil Moisture", "Soil Temperature", "MET Variable"],
    value="Soil Moisture",
    description="Plot Type:"
)

# Station selector
station_selector = widgets.Dropdown(
    options=[1, 2, 3, 4, 5, 6],
    value=1,
    description="Station:"
)

# Year selector
year_selector = widgets.Dropdown(
    options=list(range(2015, 2022)),
    value=2015,
    description="Year:"
)

# Month selector
month_selector = widgets.Dropdown(
    options=list(range(1, 13)),
    value=1,
    description="Month:"
)

# MET Variable selector
met_variable_selector = widgets.Dropdown(
    options=['Tair', 'RH', 'Wind speed', 'Srad'],
    value='Tair',
    description="MET Var:"
)

# Time resolution selector
time_resolution_selector = widgets.Dropdown(
    options=["Monthly", "Yearly"],
    value="Monthly",
    description="Time Type:"
)

# Disable the month selector if time resolution is Yearly.
def update_month_visibility(*args):
    if time_resolution_selector.value == "Yearly":
        month_selector.disabled = True
    else:
        month_selector.disabled = False

time_resolution_selector.observe(update_month_visibility, names="value")
update_month_visibility()  

# Create an interactive output area.
out_fig = interactive_output(update_plot, {
    'plot_type': plot_type_selector,
    'station_id': station_selector,
    'year': year_selector,
    'month': month_selector,
    'met_variable': met_variable_selector,
    'time_resolution': time_resolution_selector
})

# Layout the controls and output.
ui = VBox([
    HBox([plot_type_selector, station_selector, year_selector, month_selector, met_variable_selector, time_resolution_selector])
])

display(ui, out_fig)
```
