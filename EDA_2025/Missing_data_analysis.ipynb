```python
# import libraries
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import ipywidgets as widgets
from ipywidgets import interact
from pathlib import Path
from ipywidgets import VBox, HBox, Button, interactive_output

def load_revised_data(station_id):
    """
    Load revised final data from CSV.
    
    Args:
        station_id (int): Station number (1-6)
        
    Returns:
        pd.DataFrame
    """
    base_dir = Path("../EDA_2025/cleaned_data")
    file_path = base_dir / f"Station{station_id}_cleaned_data.csv"
    df = pd.read_csv(file_path, index_col=0, parse_dates=True)
    return df
```

```python
def plot_soil_moisture_interactive(station_id=1, year=2015, month=1):
    # Load revised data for the selected station
    df = load_revised_data(station_id)
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year and month
    df_filtered = df[(df.index.year == year) & (df.index.month == month)]
    
    if df_filtered.empty:
        print(f"No soil moisture data available for Station {station_id} in {year}-{month:02d}")
        return
    
    # Define the soil moisture columns and colors
    swc_cols = ['SWC_5', 'SWC_10', 'SWC_20', 'SWC_50']
    colors = {
        'SWC_5': '#4682B4',  # blue
        'SWC_10': '#FFA500',  # orange
        'SWC_20': '#3CB371',  # green
        'SWC_50': '#DC143C'   # red
    }
    
    # Create the Plotly figure
    fig = go.Figure()
    for col in swc_cols:
        fig.add_trace(go.Scatter(
            x=df_filtered.index,
            y=df_filtered[col],
            mode='lines',
            name=col,
            line=dict(color=colors[col])
        ))
    
    fig.update_layout(
        title=f"Soil Moisture for Station {station_id} in {year}-{month:02d}",
        xaxis_title="Date",
        yaxis_title="Soil Moisture (m³/m³)",
        width=1000,
        height=500
    )
    fig.show()

def plot_soil_moisture_yearly(station_id=1, year=2015):
    """
    Plot soil moisture data for the full selected year.
    
    Args:
        station_id (int): The station number.
        year (int): The selected year.
    """
    # Load revised data for the selected station
    df = load_revised_data(station_id)
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year
    df_filtered = df[df.index.year == year]
    
    if df_filtered.empty:
        print(f"No soil moisture data available for Station {station_id} in {year}")
        return
    
    # Define the soil moisture columns and colors
    swc_cols = ['SWC_5', 'SWC_10', 'SWC_20', 'SWC_50']
    colors = {
        'SWC_5': '#4682B4',   # blue
        'SWC_10': '#FFA500',  # orange
        'SWC_20': '#3CB371',  # green
        'SWC_50': '#DC143C'   # red
    }
    
    # Create the Plotly figure
    fig = go.Figure()
    for col in swc_cols:
        fig.add_trace(go.Scatter(
            x=df_filtered.index,
            y=df_filtered[col],
            mode='lines',
            name=col,
            line=dict(color=colors[col])
        ))
    
    fig.update_layout(
        title=f"Soil Moisture for Station {station_id} in {year}",
        xaxis_title="Date",
        yaxis_title="Soil Moisture (m³/m³)",
        width=1000,
        height=500
    )
    fig.show()

def plot_soil_temperature_interactive(station_id=1, year=2015, month=1):
    # Load revised data for the selected station
    df = load_revised_data(station_id)
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year and month
    df_filtered = df[(df.index.year == year) & (df.index.month == month)]
    
    if df_filtered.empty:
        print(f"No soil temperature data available for Station {station_id} in {year}-{month:02d}")
        return
    
    # Define the soil temperature columns and colors
    temp_cols = ['T_5', 'T_10', 'T_20', 'T_50']
    colors = {
        'T_5': '#4682B4',  # blue
        'T_10': '#FFA500',  # orange
        'T_20': '#3CB371',  # green
        'T_50': '#DC143C'   # red
    }
    
    # Create the Plotly figure
    fig = go.Figure()
    for col in temp_cols:
        fig.add_trace(go.Scatter(
            x=df_filtered.index,
            y=df_filtered[col],
            mode='lines',
            name=col,
            line=dict(color=colors[col])
        ))
    
    fig.update_layout(
        title=f"Soil Temperature for Station {station_id} in {year}-{month:02d}",
        xaxis_title="Date",
        yaxis_title="Temperature (°C)",
        width=1000,
        height=500
    )
    fig.show()

def plot_soil_temperature_yearly(station_id=1, year=2015):
    """
    Plot soil temperature data for the full selected year.
    
    Args:
        station_id (int): The station number.
        year (int): The selected year.
    """
    # Load revised data for the selected station
    df = load_revised_data(station_id)
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year
    df_filtered = df[df.index.year == year]
    
    if df_filtered.empty:
        print(f"No soil temperature data available for Station {station_id} in {year}")
        return
    
    # Define the soil temperature columns and colors
    temp_cols = ['T_5', 'T_10', 'T_20', 'T_50']
    colors = {
        'T_5': '#4682B4',   # blue
        'T_10': '#FFA500',  # orange
        'T_20': '#3CB371',  # green
        'T_50': '#DC143C'   # red
    }
    
    # Create the Plotly figure
    fig = go.Figure()
    for col in temp_cols:
        fig.add_trace(go.Scatter(
            x=df_filtered.index,
            y=df_filtered[col],
            mode='lines',
            name=col,
            line=dict(color=colors[col])
        ))
    
    fig.update_layout(
        title=f"Soil Temperature for Station {station_id} in {year}",
        xaxis_title="Date",
        yaxis_title="Temperature (°C)",
        width=1000,
        height=500
    )
    fig.show()

def plot_met_variable_interactive(station_id, variable='Tair', year=2015, month=1):
    """
    Plot a MET variable for a specific station, year, and month,
    and include the station's Latitude and Longitude (if available) in the title.
    
    Args:
        station_id (int): The MET station number.
        variable (str): The variable to plot (e.g., 'Tair', 'RH', 'Wind speed', 'Srad').
        year (int): The selected year.
        month (int): The selected month.
    """
    # Load the MET data for the selected station (using your revised data loader)
    df = load_revised_data(station_id)  # or load_cleaned_data if using that function for MET revised files
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year and month
    df_filtered = df[(df.index.year == year) & (df.index.month == month)]
    
    if df_filtered.empty:
        print(f"No meteorological data available for Station {station_id} in {year}-{month:02d}")
        return go.Figure()  # Return an empty figure for consistency
    
    # Get the station location from the first row if available
    if 'Latitude' in df_filtered.columns and 'Longitude' in df_filtered.columns:
        lat = df_filtered['Latitude'].iloc[0]
        lon = df_filtered['Longitude'].iloc[0]
        location_info = f" (Lat: {lat}, Lon: {lon})"
    else:
        location_info = ""
    
    # Create the Plotly figure
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df_filtered.index,
        y=df_filtered[variable],
        mode='lines',
        name=variable
    ))
    
    fig.update_layout(
        title=f"{variable} for MET Station {station_id}{location_info} in {year}-{month:02d}",
        xaxis_title="Date",
        yaxis_title=variable,
        width=1000,
        height=500
    )
    fig.show()

def plot_met_variable_yearly(station_id=1, variable='Tair', year=2015):
    """
    Plot the full-year time series for a specified MET variable.
    
    Args:
        station_id (int): The MET station number.
        variable (str): The variable to plot ('Tair', 'RH', 'Windspeed', 'Srad').
        year (int): The selected year.
    """
    # Load the MET data for the selected station
    df = load_revised_data(station_id)  # or load_cleaned_data if that is used for MET revised files
    df.index = pd.to_datetime(df.index)
    
    # Filter data for the selected year
    df_filtered = df[df.index.year == year]
    
    if df_filtered.empty:
        print(f"No {variable} data available for MET Station {station_id} in {year}")
        return go.Figure()  # Return an empty figure if there's no data
    
    # Get the station location if available
    if 'Latitude' in df_filtered.columns and 'Longitude' in df_filtered.columns:
        lat = df_filtered['Latitude'].iloc[0]
        lon = df_filtered['Longitude'].iloc[0]
        location_info = f" (Lat: {lat}, Lon: {lon})"
    else:
        location_info = ""
    
    # Create the Plotly figure
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df_filtered.index,
        y=df_filtered[variable],
        mode='lines',
        name=variable,
        line=dict(width=2)
    ))
    
    fig.update_layout(
        title=f"{variable} for MET Station {station_id}{location_info} in {year}",
        xaxis_title="Date",
        yaxis_title=variable,
        width=1000,
        height=500
    )
    fig.show()

def update_plot(plot_type, station_id, year, month, met_variable, time_resolution):
    if plot_type == "Soil Moisture":
        if time_resolution == "Monthly":
            fig = plot_soil_moisture_interactive(station_id, year, month)
        else:
            fig = plot_soil_moisture_yearly(station_id, year)
    elif plot_type == "Soil Temperature":
        if time_resolution == "Monthly":
            fig = plot_soil_temperature_interactive(station_id, year, month)
        else:
            fig = plot_soil_temperature_yearly(station_id, year)
    elif plot_type == "MET Variable":
        if time_resolution == "Monthly":
            fig = plot_met_variable_interactive(station_id, met_variable, year, month)
        else:
            fig = plot_met_variable_yearly(station_id, met_variable, year)
    else:
        fig = go.Figure()
    return fig

#############################################
# Create Widgets for the Interface
#############################################
plot_type_selector = widgets.Dropdown(
    options=["Soil Moisture", "Soil Temperature", "MET Variable"],
    value="Soil Moisture",
    description="Plot Type:"
)

station_selector = widgets.Dropdown(
    options=[1, 2, 3, 4, 5, 6],
    value=1,
    description="Station:"
)

year_selector = widgets.Dropdown(
    options=list(range(2015, 2022)),
    value=2015,
    description="Year:"
)

month_selector = widgets.Dropdown(
    options=list(range(1, 13)),
    value=1,
    description="Month:"
)

met_variable_selector = widgets.Dropdown(
    options=['Tair', 'RH', 'Windspeed', 'Srad'],
    value='Tair',
    description="MET Var:"
)

time_resolution_selector = widgets.Dropdown(
    options=["Monthly", "Yearly"],
    value="Monthly",
    description="Time Type:"
)

# Disable the month selector if time resolution is Yearly.
def update_month_visibility(*args):
    if time_resolution_selector.value == "Yearly":
        month_selector.disabled = True
    else:
        month_selector.disabled = False

time_resolution_selector.observe(update_month_visibility, names="value")
update_month_visibility()  # Initialize based on default value

#############################################
# Create an Interactive Output Area
#############################################
out_fig = interactive_output(update_plot, {
    'plot_type': plot_type_selector,
    'station_id': station_selector,
    'year': year_selector,
    'month': month_selector,
    'met_variable': met_variable_selector,
    'time_resolution': time_resolution_selector
})

ui = VBox([
    HBox([plot_type_selector, station_selector, year_selector, month_selector, met_variable_selector, time_resolution_selector])
])

#############################################
# Display the Interface and Plot
#############################################
display(ui, out_fig)
```

## Imputation Strategy for Missing SWC Data
Need to: Classify short, medium, long, and very large gaps, with an assessment of using Station 3 (which has complete data) as a reference.

1. **Short Gaps** (Hours to Days) (<24 hours)
Straightforward interpolation and local data can recover the missing values:
- Statistical Interpolation: Fill the gap by interpolating between the nearest known values.
- Moving Averages: As a quick check, take the average of a few data points before and after the missing interval to estimate the gap.
- Reference to Nearby Data: If Station 3 (or another nearby station) has data for the same period, use it as a sanity check or backup.

2. **Medium Gaps** (Days to Weeks) (1–7 days) (24–168 hours)
Simple interpolation may be inadequate. Here, we leverage time-series models and relationships with other variables:
- Time-Series Models (ARIMA) (Alternative； Kalman filtering may also good in this case)
- Regression with Meteorological Variables: Build a regression model to predict SWC using related environmental factors (and possibly other station data). Soil moisture is strongly driven by factors like precipitation , temperature, humidity, wind, and soil characteristics. Maybe use met data to predict.
- Short-Term Climate Analogues: If time-series modeling is difficult (say the gap is too close to the start of data or there’s an anomaly), consider using data from a similar period as a proxy.

Challenges & Considerations: The key difficulty is capturing any sudden events within the gap – e.g. an unseen heavy rainfall or irrigation that dramatically changed soil moisture. Time-series models might miss that unless rainfall data is included. 

To do: Cross-verify the model’s output against any clues in other data (neighboring stations or records of rain). It’s also wise to compare multiple methods: for example, fill the gap with both an ARIMA model and a regression model and see if they agree reasonably.


3. **Long Gaps** (Weeks to Months) (7–30 days) (168 hours–720 hours)
In this case, the continuity in the data is severely broken, and more advanced imputation techniques or external data sources are needed. These longer gaps require leveraging historical patterns and data from other stations or employing machine learning to reconstruct the missing period:
- Machine Learning Models: (1)Random Forests (2)XGBoost (3)neural networks
Example: a Random Forest could be trained on Station 1’s SWC as the target, with features like Station 3’s SWC, rainfall, temperature, day of year (to capture seasonality), etc., using all the days where Station 1 has data. Once trained, feed it the features for the dates inside the gap to estimate the missing SWC. Machine learning can capture interactions (e.g., how the effect of rainfall on SWC might depend on temperature) better than simple linear regression.
Consideration: (1) Need a significant amount of training data and assume that past patterns hold during the missing period.
If the gap occurs during an unprecedented event (say an unusually wet or dry season not seen in the training data), the model might not predict it well.
(2) There’s also a risk of overfitting – the model might fit the known data too closely and not generalize to the gap.

To do: To mitigate this, use techniques like cross-validation (e.g., train on one year of data to predict a different year’s known data) to ensure the model is robust. Additionally, feature engineering is important: include seasonal indicators or cumulative rainfall over a preceding period as inputs, etc., to help the model infer the state of soil moisture.

- Historical/Seasonal Averages: Incorporate seasonal patterns into the estimation. If a station has multiple years of data, compute the typical soil moisture for each time of year (monthly or even bi-weekly averages). For a gap of weeks or months, these seasonal normals can serve as a baseline to build upon.
  
- Hybrid Approaches: maybe hard to do.

4. **Very Large Gaps** (Months to Years) (over 30 days) (>720 hours)
- Seasonal Climatology Estimates: This means inferring values from the station’s typical seasonal cycle.
For instance, for Station 4 SWC_50, you could compute the average annual pattern of 50 cm soil moisture (perhaps using Station 4’s other years or even Station 3’s 50 cm data as a proxy if Station 4 has no history at that depth). Then populate the missing year with those typical values for each period. This will at least give a dataset that “looks” seasonally realistic (wetter during wet season, drier in dry season).
