```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20, 10]
import pandas as pd
%matplotlib inline
```

```python
# Read and set the index to be the date

df = pd.read_csv('../datasets/TX-Data/soil_station/SM_1.dat', sep=",", parse_dates=["Date"], index_col="Date")

# pandas remove spaces in the column names
df.columns = df.columns.str.replace(' ','')
df['SWC_5'] = df['SWC_5'].astype(float)
df['SWC_10'] = df['SWC_10'].astype(float)
df['SWC_20'] = df['SWC_20'].astype(float)
df['SWC_50'] = df['SWC_50'].astype(float)

df['T_5'] = df['T_5'].astype(float)
df['T_10'] = df['T_10'].astype(float)
df['T_20'] = df['T_20'].astype(float)
df['T_50'] = df['T_50'].astype(float)
df
```

```python
# Docs here https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html#pandas.DataFrame.plot
%matplotlib inline


df.loc["2015":"2022"].SWC_5.resample('D').mean(numeric_only=True).plot(figsize=(12,6))
plt.savefig("SWC_5-Station-1-All-Years.pdf", format="pdf", bbox_inches="tight")

plt.show()
```

```python
# Docs here https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html#pandas.DataFrame.plot
%matplotlib inline
years = [str(item) for item in range(2015, 2022)]

for i in years:
    df.loc[i].SWC_5.resample('D').mean(numeric_only=True).plot(figsize=(12,6))
    plt.savefig("SWC_5-Station-1-Year"+i+".pdf", format="pdf", bbox_inches="tight")

    plt.show()
```

```python
%matplotlib inline
years = [str(item) for item in range(2015, 2022)]

for i in years:
    df.loc[i].resample('D').mean(numeric_only=True).plot(figsize=(12,6), y=["SWC_5", "SWC_10", "SWC_20", "SWC_50"])
    plt.savefig("SWC_5_10_20_50_-Station-1-Year"+i+".pdf", format="pdf", bbox_inches="tight")
    plt.show()
```

```python
%matplotlib inline
years = [str(item) for item in range(2015, 2022)]

for i in years:
    df.loc[i].resample('D').mean(numeric_only=True).plot(figsize=(12,6), y=["T_5", "T_10", "T_20", "T_50"])
    plt.savefig("T_5_10_20_50_-Station-1-Year"+i+".pdf", format="pdf", bbox_inches="tight")

    plt.show()
```
