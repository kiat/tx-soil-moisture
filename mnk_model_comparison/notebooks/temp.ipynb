```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load results
df = pd.read_csv("old/all_results.csv")
df['model'] = df['model'].str.upper()

# Plot for 48-hour horizon
df_48 = df[df['output_horizon'] == 48]

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_48, x='model', y='rmse', palette='Set2')
plt.title('RMSE by Model @ 48-Hour Forecast Horizon')
plt.xlabel('Model')
plt.ylabel('RMSE')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('old/all_results.csv')

# Optional: filter to only models of interest
models_to_plot = ['arima', 'arimax', 'lstm', 'bilstm', 'cnn', 'xgboost']
df = df[df['model'].str.lower().isin(models_to_plot)]

# Normalize model names for clean plotting
df['model'] = df['model'].str.upper()

# Get list of unique seasons
seasons = df['season'].dropna().unique()

# Plot one figure per season
for season in seasons:
    df_season = df[df['season'] == season]
    
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df_season, x='model', y='rmse', palette='Set2')
    plt.title(f'RMSE Comparison by Model - {season.capitalize()}')
    plt.xlabel('Model')
    plt.ylabel('RMSE')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
