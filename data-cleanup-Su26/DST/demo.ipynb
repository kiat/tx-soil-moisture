```python
import pandas as pd

from dup_cleaner import dup_cleaner
from time_cleaner import time_cleaner
```

```python
def read_soil(raw_file):

    df = pd.read_csv(raw_file, header=5, dtype = str)

    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

    # make the rest floats
    for col in df.columns:
        if col != 'Date' and col != 'Flag':
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # sort timestamps in case they are not in order
    df = df.sort_values(by='Date')

    return df

def read_met(raw_file):

    df = pd.read_csv(raw_file, header = 6, dtype = str)

    df = df.drop(df.index[0:2])

    df = df.drop(columns=['RECORD'])

    df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'], errors='coerce')

    # make the rest floats
    for col in df.columns:
        if col != 'TIMESTAMP':
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # sort timestamps in case they are not in order
    df = df.sort_values(by='TIMESTAMP')

    return df
```

```python
soil_path = r"C:\pyt\tx-soil-moisture\datasets\TxSON_data_2026-02-24\CB01.dat"
met_path = r"C:\pyt\tx-soil-moisture\datasets\TxSON_data_2026-02-24\CB04_met.dat"

soil_df = read_soil(soil_path)
met_df = read_met(met_path)
```
