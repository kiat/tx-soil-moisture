```python
from scripts.gap_report import *
import pandas as pd
import os
```

```python
data = r'.\prewashed_data'
gap_report_path = r'.\gap_reports'

if not os.path.exists(gap_report_path):
    os.makedirs(gap_report_path)

for file in os.listdir(data):
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(data, file), parse_dates=[0], index_col=0)
        report = gap_report(df)
        report = report.reset_index()
        report.to_csv(os.path.join(gap_report_path, f"{file}_gap_report.csv"), index=False)
```
